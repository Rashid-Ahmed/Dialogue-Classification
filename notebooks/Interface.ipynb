```python
import pandas as pd
import torch
import os
from utils.train_model import load_existing_model
from utils.get_embeddings import get_model
from utils.networking import start_tcp_server
```

```python
SENTENCE_SIZE = 79
EMBEDDING_SIZE = 100
DATA_DIR = 'data'
EMBEDDING_MODEL_NAME = 'cc.de.100.bin'
MODEL_TYPES = ['self', 'parents', 'teacher', 're', 'cause', 'none', 'soc']
OUTPUT_SIZE = [2, 2, 2, 2, 2, 2, 6]
NUM_LAYERS = 2
HIDDEN_SIZE = 400
#device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
device = 'cpu'
embedding_model = get_model(os.path.join(DATA_DIR, EMBEDDING_MODEL_NAME))

```

```python
MODELS = []
for i in range(len(MODEL_TYPES)):
  print (MODEL_TYPES[i])
  MODELS.append(load_existing_model(os.path.join('models', 'model_'+MODEL_TYPES[i]+'.ckpt'), device, OUTPUT_SIZE[i], HIDDEN_SIZE , NUM_LAYERS, EMBEDDING_SIZE))

start_tcp_server('localhost', 10000, MODEL_TYPES, MODELS, embedding_model, EMBEDDING_SIZE, SENTENCE_SIZE, device)
```
